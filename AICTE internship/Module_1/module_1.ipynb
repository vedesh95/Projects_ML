```python
tdf["CP"]=pd.DataFrame(df["Close Price"])
tdf["Date"]=df["Date"]
tdf["Trade"]=df["No. of Trades"]
def vwap(grp):
    return (grp.CP*grp.Trade).sum() / grp.CP.sum()
```

### Task 1.5

```python
#get average price of last n days
def averageprice(n):
    last90d=df.iloc[494-n:,:]
    temp=last90d["Close Price"]
    return temp.mean()
```

```python
temp=averageprice(10)
temp
```

```python
#get progit or loss of last n days
def profit_or_loss(n):
    lastn=df.iloc[494-n,8]
    lastv=df.iloc[493,8]
    li=[]
    if((lastv-lastn)>=0):
        li.append("profit :")
        li.append((lastv-lastn)/lastn*100)
    else:
        li.append("loss :")
        li.append((lastn-lastv)/lastn*100)
    return li
```

```python
#last one week trend
profit_or_loss(7)
```

### Task 1.7

```python
#Convert temp dataframe to numpy array which contains DAy_Perc_Change column values
type(temp)
temp = temp.astype('float64') 
```

```python
type(temp)
li=[]
for i in range(len(temp)):
    if(temp[i]>=-0.005 and temp[i]<0.005):
        li.append("No change")
    elif(temp[i]>=0.005 and temp[i]<0.01):
        li.append("Slight positive")
    elif(temp[i]>=0.01 and temp[i]<0.03):
        li.append("Positive")
    elif(temp[i]>=0.03 and temp[i]<0.07):
        li.append("Among top gainers")
    elif(temp[i]>=0.07):
        li.append("Bull run")
    elif(temp[i]>-0.01 and temp[i]<-0.005):
        li.append("Slight Negative")
    elif(temp[i]>-0.03 and temp[i]<=-0.01):
        li.append("Negative")
    elif(temp[i]>-0.07 and temp[i]<=-0.03):
        li.append("Among top losers")
    elif(temp[i]<-0.07):
        li.append("Bear Drop")
```

```python
df["Trend"]=li
```
